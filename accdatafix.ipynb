```python
import csv 
import pandas as pd


def Go_csv(file_name):
    
    with open(file_name, newline='') as csvfile: 
    
        datas = list(csv.reader(csvfile)) 
    
    for i in range(len(datas)):

        err=0
        m_err=True
        while(m_err):
            if(len(datas[i]) != 9):
                datas[i][6] = datas[i][6]+datas[i][7]
                del(datas[i][7])
                err = err + 1
            else:
                m_err=False
    
    df = pd.DataFrame(datas)
    
    return df
```
