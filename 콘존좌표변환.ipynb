```python
from pyproj import Proj, transform
import numpy as np
import pandas as pd
import folium
import time
```

```python
# Change coordinate system
# korea 2000/central belt 2010 (epsg:5186) to wgs84(epsg:4326)

def Merges(list1, list2):
    
    a = pd.DataFrame(list1)
    b = pd.DataFrame(list2)
    
    c = pd.merge(a,b, on = a.index)
    c.columns = range(c[c.index==0].size)
    c.set_index(c[0])
    c=c.drop(0,1)
    
    return c
```

```python
# grs80x,grs80y 각각의 리스트를 입력
# 해당하는 좌표의 경도,위도 데이터 프레임을 출력하는 함수

def Transform(xs,ys):
    
    inProj = Proj(init = 'epsg:5186')
    outProj= Proj(init = 'epsg:4326')
    
    nys=[]
    nxs=[]
    
    now=0
    tmp=0
    
    for i in range(len(xs)):
        x = xs[i]
        y = ys[i]
        #<grs80x>391265.163647</grs80x>
        #<grs80y>297630.026461</grs80y>
        nx,ny = transform(inProj,outProj,x,y)     # 새로운 좌표계 
        
        now = int(i/len(xs)*100) # 진행상황 %
        if(now>tmp): # 중복된 %는 출력 안함
            print('%3d%%' % now, end='', flush = True)
        
        tmp = now 
        
        nxs.append(nx)
        nys.append(ny)
     
    print('')
    return Merges(nys,nxs)
```

```python
df1 = pd.read_csv('conzone1.csv',sep=',',encoding='CP949')
df1 = df1.drop('Unnamed: 0',1)
df1.tail(5)
```

```python


start=time.time()

df2 = Transform(df1['grs80x'],df1['grs80y'])

end = time.time()

print('\n%d초 걸림'%(int(end-start)))


df2.tail(5)
```

```python
col_v = df1.columns.values
col_v = np.append(col_v,['위도','경도'])

df3= Merges(df1,df2)
df3.columns=col_v
df3.tail(5)
```

```python
df3.to_csv(r'acc_conzone.csv',encoding="CP949")
```
