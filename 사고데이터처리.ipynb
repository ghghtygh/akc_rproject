```python
#df1 = pd.read_csv('acc170715.csv',sep="\t",encoding='CP949')
import pandas as pd
import numpy as np
import csv

def Open_csv(file_name):
     
    df1 = pd.read_csv(file_name, sep="\t", encoding='CP949', error_bad_lines=False, warn_bad_lines=False)
    df1.index = range(len(df1))
    df1.columns=range(1)
    
    df2 = df1[0].str.split(',',n=11,expand=True)
    
    #df2 = pd.read_csv(file_name,sep=',',encoding='CP949',error_bad_lines=False)
    
    return df2

def Go_csv(file_name):
    
    with open(file_name, errors='ignore', newline='') as csvfile: 
    
        datas = list(csv.reader(csvfile)) 
    
    for i in range(len(datas)):

        err=0
        m_err=True
        while(m_err):
            if(len(datas[i]) != 9):
                datas[i][6] = datas[i][6]+datas[i][7]
                del(datas[i][7])
                err = err + 1
            else:
                m_err=False
    
    df = pd.DataFrame(datas)
    
    return df
```

```python

#디렉토리에 있는 파일명 불러오는 소스

import os

#path = r"C:\Users\정환\Desktop\bigdatafile\team\수집\data"
path = r"Ac_Day_20142019"
folder='Ac_Day_20142019'

df3 = pd.DataFrame()
file_list = os.listdir(path)

# 모든 csv 파일 읽고 한 데이터 프레임에 저장

import time

start = time.time()

# 정상/오류 파일 개수 
count = 0
error_count=0

#진행 상황
m_sum = 0
tmp = -1
m_all = 0



for i in file_list:
    
    strs = folder+'/'+i
    m_stat = os.stat(str)
    m_size=m_stat.st_size
    m_sum+=m_size

m_all = m_sum
m_sum = 0

print(folder,' : %.2fMB' % (m_all/(1024*1024)))

for i in file_list:

    strs = folder+'/'+i
    
    try:
        df3=df3.append(Open_csv(strs))
    except:
        print(strs," 파일에서 에러발생")
        error_count+=1
    
    
    count+=1
        
    m_stat = os.stat(strs)
    m_size=m_stat.st_size
    m_sum+=m_size
    now = int(m_sum/m_all*100)
    if(now>tmp):
        print('%3d%%' % now, end='', flush = True)
    tmp = now

end = time.time()

print('\n소요시간 : %d초' % int(end-start))
print('총 파일 %6d개' % count)
print('총 에러 %6d개' % error_count)
print('총 사고 %6d건' % len(df3[df3[4]=='1']))

#==> 소요시간 : 초
#==> 총 파일 1577개
#==> 총 에러 0개
#==> 총 사고 265882건

#df3.columns=("생성일", "사고ID" , "발생시간", "상태", "분류코드", "분류명", "메세지", "coodX", "coodY")
```
